```python
import binotools as bt
from astropy.io import fits
import datetime, time

def find_data():
    """
    Finds raw data
    """
    datadir    = "/Users/gonzalo/Desktop/Code/Gonzalo_Binospec/code/data/large_files/Raw/mask2/Expositions/"
    files=sorted(bt.insensitive_glob(datadir+'/*/obj_clean*abs_slits_lin.fits'))
    return files

def replaceParamRC(lines,paramName,value):
    for i,l in enumerate(lines):
        if l.split("=")[0]==paramName:
            sep = l.split("=")
            sep[1]=str(value)+"\n"
            lines[i]=sep[0]+'='+sep[1]
    return lines

def writeRC(lines,dir,tipo):
    f = open(dir[:-5]+"_"+tipo+'.rc', "w")
    f.writelines(lines)
    f.close()
   
    
def replaceFileSOF(lines,paramName,value):

    for i,l in enumerate(lines):
        if paramName in l.split(" ")[1]:
            sep = l.split(" ")
            sep[0]=str(value)
            lines[i]=sep[0]+' '+sep[1]

    return lines

def writeSOF(lines,dir,tipo):
    f = open(dir[:-5]+"_"+tipo+'.sof', "w")
    f.writelines(lines)
    f.close()


def adaptRC():
    """
    Adapts the demo .rc file to what is in each exposures header
    """
    demodir   = "/Users/gonzalo/Desktop/Code/Gonzalo_Binospec/code/Bino_analysis/rc_files/"
    demoFiles={}
    for i in sorted(bt.insensitive_glob(demodir+'/Default*.rc')):
        if "model" in i:
            demoFiles["modelRC"]=i
        if "calctrans" in i:
            demoFiles["calcRC"]=i
        if "correct" in i:
            demoFiles["correctRC"]=i

    demodir   = "/Users/gonzalo/Desktop/Code/Gonzalo_Binospec/code/Bino_analysis/sof_files/"
    for i in sorted(bt.insensitive_glob(demodir+'/Default*.sof')):
        if "model" in i:
            demoFiles["modelSOF"]=i
        if "calctrans" in i:
            demoFiles["calcSOF"]=i
        if "correct" in i:
            demoFiles["correctSOF"]=i

    LinesModel = open(demoFiles["modelRC"], 'r').readlines()
    LinesCalc = open(demoFiles["calcRC"], 'r').readlines()
    LinesCorr = open(demoFiles["correctRC"], 'r').readlines()

    SOFModel = open(demoFiles["modelSOF"], 'r').readlines()
    SOFCalc = open(demoFiles["calcSOF"], 'r').readlines()
    #SOFCorr = open(demoFiles["correctRC"], 'r').readlines()

    for ss,spec2D_file in enumerate(find_data()):
        spec2D_hdu = fits.open(spec2D_file)
        hdr = spec2D_hdu[0].header
        UTC = time.strptime(hdr['UT'],'%H:%M:%S')
        UTC_sec = datetime.timedelta(hours=UTC.tm_hour,minutes=UTC.tm_min,seconds=UTC.tm_sec).total_seconds()

        LinesModel = replaceParamRC(LinesModel,'UTC_VALUE',UTC_sec)
        LinesModel = replaceParamRC(LinesModel,'OBSERVING_DATE_VALUE',hdr['MJD'])
        LinesModel = replaceParamRC(LinesModel,'TELESCOPE_ANGLE_VALUE',hdr['EL'])
        LinesModel = replaceParamRC(LinesModel,'PRESSURE_VALUE',hdr['ADCPRES'])
        LinesModel = replaceParamRC(LinesModel,'TEMPERATURE_VALUE',hdr['ADCTEMP'])
        LinesModel = replaceParamRC(LinesModel,'MIRROR_TEMPERATURE_VALUE',hdr['ADCTEMP'])
        
        Num=spec2D_file.split("_")[3]
        dir=spec2D_file.split("/")[1:-1]
        dir="/"+'/'.join([d for d in dir])+"/"
        
        SOFModel = replaceFileSOF(SOFModel,'SCIENCE',spec2D_file)
        SOFCalc = replaceFileSOF(SOFCalc,'SCIENCE',spec2D_file,'calctrans')
        SOFCalc = replaceFileSOF(SOFCalc,'MODEL_MOLECULES',dir+Num+"/MODEL_MOLECULES.fits",'calctrans')
        SOFCalc = replaceFileSOF(SOFCalc,'ATM_PARAMETERS',dir+Num+"/ATM_PARAMETERS.fits",'calctrans')
        SOFCalc = replaceFileSOF(SOFCalc,'BEST_FIT_PARAMETERS',dir+Num+"/BEST_FIT_PARAMETERS.fits",'calctrans')

        writeRC(LinesModel,spec2D_file,"model")
        writeRC(LinesCalc,spec2D_file,"calc")
        writeRC(LinesCorr,spec2D_file,"correct")

        writeSOF(SOFModel,spec2D_file,"model")
    



        
        
adaptRC()

```
