```python
import numpy as np
from astropy.io import ascii
```

```python
Catalog=ascii.read("../A-Catalogs/Binospec-Candels.cat")
Catalog=Catalog[Catalog["Type"]=="LAE"]
Catalog.sort("z")
for gal in Catalog:
    ewlim=""
    if np.isnan(gal["zsys"]) and not np.isnan(gal["EWHa"]):
        ewlim = "< "

    fesclim=""
    if np.isnan(gal["zsys"]) and not np.isnan(gal["fesc"]):
        fesclim = "> "


    ID=gal["id_charlotte"].replace("_","-")
    ID=ID.split("-")
    ID=ID[0]+"-"+ID[2] if "Stark" in ID[0] else "-".join(ID)
    ID=ID.split("-")
    ID=ID[0]+"-"+ID[3] if "Jung" in ID[0] else "-".join(ID)
    ID=ID.split("-")
    ID=ID[0]+"-"+ID[2] if "Hu" in ID[0] else "-".join(ID)


    zsys = np.round(gal["zsys"], 4)
    zsys = f"{zsys:.4f}" if not np.isnan(gal["zsys"]) else "--"

    z=np.round(gal["z"],4)
    z=f"{z:.4f}" if not np.isnan(gal["z"]) else "--"

    ra=np.round(gal["ra_candels"],6)
    ra=f"{ra:.6f}" if not np.isnan(gal["ra_candels"]) else "--"
    dec=np.round(gal["dec_candels"],6)
    dec=f"{dec:.6f}" if not np.isnan(gal["dec_candels"]) else "--"

    Muv=np.round(gal["Muv"],1)
    Muv=f"{Muv:.1f}" if not np.isnan(Muv) else "--"

    dMuv=np.round(gal["dMuv"],1)
    dMuv=" $\pm$ "+ f"{dMuv:.1f}" if not np.isnan(dMuv) else ""

    slope=np.round(gal["slope"],1)
    slope=f"{slope:.1f}" if not np.isnan(slope) else "--"
    dslope=np.round(gal["dslope"],1)
    dslope=" $\pm$ "+f"{dslope:.1f}" if not np.isnan(dslope) else ""

    EWHa=np.round(gal["EWHa"],0)
    EWHa=int(EWHa) if not np.isnan(EWHa) else "--"
    dEWHa=np.round(gal["dEWHa"],0)
    dEWHa=" $\pm$ "+str(int(dEWHa)) if not np.isnan(dEWHa) else ""






    voff=np.round(gal["voff"],0)
    voff = int(voff) if not np.isnan(voff) else "--"
    dvoff=np.round(gal["dvoff"],0)
    dvoff=" $\pm$ "+str(int(dvoff)) if not np.isnan(dvoff) else ""

    EWLya=np.round(gal["EWLya"])
    EWLya=int(EWLya) if not np.isnan(EWLya) else "--"
    dEWLya=np.round(gal["dEWLya"])
    dEWLya=" $\pm$ "+str(int(dEWLya)) if not np.isnan(dEWLya) else ""

    FWHM=np.round(gal["LyaFWHM_int"])
    FWHM=int(FWHM) if not np.isnan(FWHM) else "--"
    dFWHM=np.round(gal["dLyaFWHM_int"])
    dFWHM=" $\pm$ "+str(int(dFWHM)) if not np.isnan(dFWHM) else ""

    
    FWHMHa= np.round( np.sqrt(gal["HaFWHM"]**2 - 180**2)     )     
    FWHMHa=int(FWHMHa) if not np.isnan(FWHMHa) else "--"
    dFWHMHa=np.round(gal["dHaFWHM"])
    dFWHMHa=" $\pm$ "+str(int(dFWHMHa)) if not np.isnan(dFWHMHa) else ""

    fesc=np.round(gal["fesc"],2)
    fesc=f"{fesc:.2f}" if not np.isnan(fesc) else "--"
    dfesc=np.round(gal["dfesc"],2)
    dfesc=" $\pm$ "+f"{dfesc:.2f}" if not np.isnan(dfesc) else ""

    voff=np.round(gal["voff"],0)
    voff = int(voff) if not np.isnan(voff) else "--"
    dvoff=np.round(gal["dvoff"],0)
    dvoff=" $\pm$ "+str(int(dvoff)) if not np.isnan(dvoff) else ""

    print(str(ID)+" & "+str(ra)+" & "+str(dec)+" & "+str(z)+" & "+str(zsys)+" & "+str(Muv)+dMuv+" & "+str(slope)+dslope 
          +" & "+str(EWLya) + dEWLya
          +" & "+str(FWHM) + dFWHM
          +" & "+fesclim+str(fesc)+dfesc+" & "+str(voff)+dvoff+"  \\\\")
```
