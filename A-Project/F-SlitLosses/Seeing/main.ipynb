```python
import glob
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
import numpy as np
from natsort import natsorted


AllMasks=["Mask1"]

for mask in AllMasks:
    dir="/Users/gonzalo/Desktop/Code/Gonzalo_Binospec/A-Project/A-Catalogs/data/large_files/Raw/"+mask+"/*/*"
    file_list = glob.glob(dir, recursive=True)
    file_list = natsorted(file_list)
    Seeings=[]
    files=[]
    exptime=[]
    for i,file in enumerate(file_list):
        fit=fits.open(file)
        seeing_comment=fit[0].header["COMMENT"]
        exptime.append(fit[0].header["EXPTIME"])
        mask = [True if "f_" in i.header["SLITOBJ"] else False for i in fit[1:]]
        fitss=np.array(fit[1:])[mask]
        temp=[]
        for fit in fitss:
            MeasurePoints=[1401,1646,1823,2023,2237,2644,2670,3085]
            Data2D=fit.data
            center=int(np.round(fit.header['SLITYPIX'],0))

            Data1D=[Data2D.T[i] for i in  MeasurePoints]
            Data1D = [np.nan_to_num(data) for data in Data1D]
            

            #plt.plot(Data1D[0])
            #plt.axvline(x=center, color='r', linestyle='--')
            #plt.show()

            def gaussian(x, amplitude, mean, stddev,C):
                return amplitude * np.exp(-(((x - mean) / stddev) ** 2)/2)+C

            fit_results = []
            for dataset in Data1D:
                x = np.arange(len(dataset))
                p0 = [np.max(dataset), center, 1,0]  # Initial guess for the parameters
                params, _ = curve_fit(gaussian, x, dataset, p0=p0)
                fit_results.append(params)
                #plt.plot(x, dataset)
                #plt.plot(x, gaussian(x, *params), 'r--')
                #plt.title(str(params[2]*2.355*0.24))
                #plt.show()
            
            median_std = np.nanmedian([params[2] for params in fit_results])
            #print(median_std)
            fwhm = 2 * np.sqrt(2 * np.log(2)) * median_std*0.24
            #print(fwhm)
            seeing=fwhm
            temp.append(seeing)
        Seeings.append(temp)
        files.append(file)

    # Now you can work with the fits_data list which contains the loaded FITS files


Seeings=np.median(Seeings,axis=1)
excluded=[]
s=[]
time=0
for i,j,e in zip(Seeings,files,exptime):
    print(e,i,j)
    if i<1.2:
        s.append(i)
        time+=e
    else:
        excluded.append(i)

```
